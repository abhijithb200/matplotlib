```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.svm import LinearSVC,SVC
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/ReubenJoe/SVM-Analysis/main/Dataset/voice.csv")
data.head()
```

```python
encoder = LabelEncoder()
data['label'] = encoder.fit_transform(data['label'])
data.head()
```

```python
 x = data.drop('label',axis=1)
 y = data.label
```

```python
scaler = StandardScaler()
scaler.fit(x)
x = scaler.transform(x)

```

```python
train_x,test_x,train_y,test_y = train_test_split(x,y,test_size=0.3,random_state=40)
```

```python
modal1 = LinearSVC()
modal1.fit(train_x,train_y)
print("Accuracy :",modal1.score(test_x,test_y))
```

```python
modal2 = SVC(kernel='sigmoid')
modal2.fit(train_x,train_y)
print("Accuracy :",modal2.score(test_x,test_y))
```

```python
modal3 = SVC(kernel='poly')
modal3.fit(train_x,train_y)
print("Accuracy :",modal3.score(test_x,test_y))
```
